## 2.x. Normalizing Pixel Values

```python
# File path to the directory containing your images
image_dir = "C:/Users/User/Documents/ML-finalproject/IFL/casia_tts_100/tp_class_100"
```

```python
# Function to normalize pixel values of images in a directory
def normalize_images(image_dir):
    # Get list of image filenames
    image_filenames = [filename for filename in os.listdir(image_dir) if filename.endswith((".jpg", ".png"))]
    
    # Initialize tqdm progress bar with total number of images
    progress_bar = tqdm(total=len(image_filenames), desc="Normalizing Images")
    
    # Loop through each image filename and its index
    for i, filename in enumerate(image_filenames, 1):
        image_path = os.path.join(image_dir, filename)
        # Load image
        image = cv2.imread(image_path)
        # Convert to float
        image = image.astype(np.float32)
        # Normalize pixel values
        image /= 255.0
        # Save normalized image
        cv2.imwrite(image_path, image * 255.0)  # Save back as uint8 (optional)
        
        # Update tqdm progress bar
        progress_bar.update(1)
        progress_bar.set_postfix({"Processed": i})
    
    # Close tqdm progress bar
    progress_bar.close()
```

```python
# Call the function to normalize images in the directory
normalize_images(image_dir)
```

## 2.x. Converting Image to Arrays

```python
# Function to load and convert images to arrays with tqdm progress bar
def load_images_as_arrays_with_tqdm(image_dir):
    images = []
    # Get list of image filenames
    image_filenames = [filename for filename in os.listdir(image_dir) if filename.endswith((".jpg", ".png"))]
    
    # Initialize tqdm progress bar with total number of images
    progress_bar = tqdm(total=len(image_filenames), desc="Loading Images")
    
    # Loop through each image filename and its index
    for i, filename in enumerate(image_filenames, 1):
        image_path = os.path.join(image_dir, filename)
        # Load image
        image = cv2.imread(image_path)
        # Convert image to array and append to the list
        images.append(image)
        
        # Update tqdm progress bar
        progress_bar.update(1)
        progress_bar.set_postfix({"Processed": i})
    
    # Close tqdm progress bar
    progress_bar.close()
    
    # Convert list of arrays to a single NumPy array
    images_array = np.array(images)
    return images_array
```

```python
# Call the function to load and convert images to arrays with tqdm
images_array = load_images_as_arrays_with_tqdm(image_dir)
```

```python
# Function to load images and corresponding ground truth masks
def load_data(tampered_dir, gt_path):
    tampered_images = []
    groundtruth_masks = []
    # Iterate through files in the directory
    for filename in os.listdir(tampered_dir):
        if filename.endswith((".jpg", ".png")):  # Assuming images are in JPG or PNG format
            # Load tampered image
            tampered_image_path = os.path.join(tampered_dir, filename)
            tampered_image = cv2.imread(tampered_image_path)
            tampered_images.append(tampered_image)
            # Load corresponding ground truth mask
            mask_filename = filename.split('.')[0] + '_gt.' + filename.split('.')[-1]
            groundtruth_mask_path = os.path.join(gt_path, mask_filename)
            groundtruth_mask = cv2.imread(groundtruth_mask_path, cv2.IMREAD_GRAYSCALE)

            # Load corresponding ground truth mask
            groundtruth_mask_path = os.path.join(gt_path, mask_filename)
            print("Ground truth mask path:", groundtruth_mask_path)  # Debugging print statement
            groundtruth_mask = cv2.imread(groundtruth_mask_path, cv2.IMREAD_GRAYSCALE)
            print("Loaded ground truth mask shape:", groundtruth_mask.shape)  # Debugging print statement
            # Ensure the mask is a 2D array
            groundtruth_mask = np.expand_dims(groundtruth_mask, axis=-1)
            # Normalize pixel values to range [0, 1]
            groundtruth_mask = groundtruth_mask / 255.0
            groundtruth_masks.append(groundtruth_mask)
    return tampered_images, groundtruth_masks
```

# 3. Modeling

## 3.1. Pre-trained U-Net Model

### from: IFL_Part01.ipynb

```python
# Define input tensor shape (512x512x3)
input_img = Input((512, 512, 3))
```

```python
def conv2d_block(input_tensor, n_filters, kernel_size=3, batchnorm=True):
    """
    A convolutional block consisting of two convolutional layers with optional batch normalization.

    Parameters:
        input_tensor (tensor): The input tensor to the convolutional block.
        n_filters (int): The number of filters (channels) in the convolutional layers.
        kernel_size (int, optional): The size of the convolutional kernel. Defaults to 3.
        batchnorm (bool, optional): Whether to apply batch normalization. Defaults to True.

    Returns:
        tensor: The output tensor after passing through the convolutional block.
    """
    # First convolutional layer
    x = Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size), kernel_initializer="he_normal",
               padding="same")(input_tensor)
    if batchnorm:
        x = BatchNormalization()(x)  # Apply batch normalization if enabled
    x = Activation("relu")(x)  # Apply ReLU activation function
    
    # Second convolutional layer
    x = Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size), kernel_initializer="he_normal",
               padding="same")(x)
    if batchnorm:
        x = BatchNormalization()(x)  # Apply batch normalization if enabled
    x = Activation("relu")(x)  # Apply ReLU activation function
    
    return x  # Return the output tensor
```

```python
def get_unet(input_img, n_filters=16, dropout=0.5, batchnorm=True):
    # contracting path
    c1 = conv2d_block(input_img, n_filters=n_filters*1, kernel_size=3, batchnorm=batchnorm)
    p1 = MaxPooling2D((2, 2)) (c1)
    p1 = Dropout(dropout*0.5)(p1)

    c2 = conv2d_block(p1, n_filters=n_filters*2, kernel_size=3, batchnorm=batchnorm)
    p2 = MaxPooling2D((2, 2)) (c2)
    p2 = Dropout(dropout)(p2)

    c3 = conv2d_block(p2, n_filters=n_filters*4, kernel_size=3, batchnorm=batchnorm)
    p3 = MaxPooling2D((2, 2)) (c3)
    p3 = Dropout(dropout)(p3)

    c4 = conv2d_block(p3, n_filters=n_filters*8, kernel_size=3, batchnorm=batchnorm)
    p4 = MaxPooling2D(pool_size=(2, 2)) (c4)
    p4 = Dropout(dropout)(p4)
    
    c5 = conv2d_block(p4, n_filters=n_filters*16, kernel_size=3, batchnorm=batchnorm)
    
    # expansive path
    u6 = Conv2DTranspose(n_filters*8, (3, 3), strides=(2, 2), padding='same') (c5)
    u6 = concatenate([u6, c4])
    u6 = Dropout(dropout)(u6)
    c6 = conv2d_block(u6, n_filters=n_filters*8, kernel_size=3, batchnorm=batchnorm)

    u7 = Conv2DTranspose(n_filters*4, (3, 3), strides=(2, 2), padding='same') (c6)
    u7 = concatenate([u7, c3])
    u7 = Dropout(dropout)(u7)
    c7 = conv2d_block(u7, n_filters=n_filters*4, kernel_size=3, batchnorm=batchnorm)

    u8 = Conv2DTranspose(n_filters*2, (3, 3), strides=(2, 2), padding='same') (c7)
    u8 = concatenate([u8, c2])
    u8 = Dropout(dropout)(u8)
    c8 = conv2d_block(u8, n_filters=n_filters*2, kernel_size=3, batchnorm=batchnorm)

    u9 = Conv2DTranspose(n_filters*1, (3, 3), strides=(2, 2), padding='same') (c8)
    u9 = concatenate([u9, c1], axis=3)
    u9 = Dropout(dropout)(u9)
    c9 = conv2d_block(u9, n_filters=n_filters*1, kernel_size=3, batchnorm=batchnorm)
    
    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = Model(inputs=[input_img], outputs=[outputs])
    return model
```

```python
# Build U-Net model with specified parameters
model_unet = get_unet(
    input_img, 
    n_filters=16, 
    dropout=0.05, 
    batchnorm=True
)

# Compile the model with appropriate loss function for binary segmentation
model_unet.compile(
    optimizer='adam', 
    loss='binary_crossentropy', 
    metrics=['accuracy']
)

# Print model summary
model_unet.summary() #Total params: 2,164,593, Trainable params: 2,161,649 (8.25 MB), Non-trainable params: 2,944 (11.50 KB)
```
