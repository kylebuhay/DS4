## 2.x. Tp: Converting Images from TIF to JPEG

```python
# Function to convert TIF files to JPEG files and overwrite original TIF files
def convert_tif_to_jpeg(directory):
    tif_files = [file for file in os.listdir(directory) if file.endswith('.tif')]
    print(f"Found {len(tif_files)} TIF files in {directory}")
    
    converted_count = 0
    for file in tqdm(tif_files, desc="Converting TIF to JPEG"):
        try:
            # Open TIF file
            tif_path = os.path.join(directory, file)
            image = Image.open(tif_path)
            
            # Convert and save as JPEG (overwriting the original TIF file)
            jpeg_path = os.path.splitext(tif_path)[0] + '.jpg'
            image.convert('RGB').save(jpeg_path, 'JPEG')
            
            # Increment converted count
            converted_count += 1
            
            # Remove original TIF file
            os.remove(tif_path)
        except Exception as e:
            print(f"Error converting {file}: {e}")
    
    print(f"Successfully converted and replaced {converted_count} TIF files with JPEG")

# Convert TIF files to JPEG in 'Tp' directory
convert_tif_to_jpeg(path+ds_path['tp'])
```

## 2.x. Tp: Resizing Images to 512 by 512

```python
# Directory containing the original images
tp_image_dir = r"C:\Users\User\Documents\ML-finalproject\CASIAv2.0\CASIA2.0_revised\Tp"

# Directory to save the resized images
tp_pp = r"C:\Users\User\Documents\ML-finalproject\IFL\casia_tts\tp_class"
```

```python
# Counter to keep track of the number of tampered images resized and saved
tp_count_resized = 0
```

```python
# Loop through each file in the tp_image_dir directory
for filename in tqdm(os.listdir(tp_image_dir), desc="Resizing Tampered Images"):
    # Construct the full path to the original image
    original_image_path = os.path.join(tp_image_dir, filename)

    # Open the image using PIL
    original_image = Image.open(original_image_path)

    # Check if image mode is RGBA
    if original_image.mode == 'RGBA':
        # Convert image to RGB format
        original_image = original_image.convert('RGB')
    elif len(original_image.getbands()) > 3:  # Check if image has more than 3 channels
        # Extract RGB channels
        r, g, b, _ = original_image.split()
        # Merge RGB channels
        original_image = Image.merge('RGB', (r, g, b))

    # Resize the image to 512x512 pixels
    resized_image = original_image.resize((512, 512))

    # Construct the full path to save the resized image in tp_pp directory
    resized_image_path = os.path.join(tp_pp, filename)

    # Save the resized image
    resized_image.save(resized_image_path)

    # Increment the counter
    tp_count_resized += 1

print(f"{tp_count_resized} images resized and saved successfully.")
```

```python
# Create a set to store unique image dimensions and channels
unique_dimensions_tp = set()

# Loop through each file in the tp_pp directory with tqdm
for filename in tqdm(os.listdir(tp_pp), desc="Checking dimensions of Tampered Images"):
    # Construct the full path to the tampered image
    image_path = os.path.join(tp_pp, filename)

    # Open the image using PIL
    image = Image.open(image_path)
    
    # Get the dimensions of the image
    width, height = image.size

    # Get the number of channels in the image
    num_channels = len(image.getbands())

    # Add the dimensions and number of channels to the set of unique dimensions
    unique_dimensions_tp.add((width, height, num_channels))

# Print the unique dimensions and channels
print("Unique dimensions and channels of tampered images:")
for dimensions in unique_dimensions_tp:
    print(f"Dimensions: {dimensions[:2]}, Channels: {dimensions[2]}")
```

# 3. Modeling

## 3.1. Pre-trained U-Net Model

### from: https://ai.stackexchange.com/questions/27776/unet-overfitting-for-binary-segmentation-of-fake-images

```python
# Define input tensor shape (512x512x3)
input_img = Input((512, 512, 3))
```

```python
def conv2d_block(input_tensor, n_filters, kernel_size=3, batchnorm=True):
    """Function to add 2 convolutional layers with the parameters passed to it"""
    # first layer
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size),
                                kernel_initializer='he_normal', padding='same')(input_tensor)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)

    # second layer
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size),
                                kernel_initializer='he_normal', padding='same')(x)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)

    return x
```

```python
def get_unet(input_img, n_filters=16, dropout=0.1, batchnorm=True):
    """Function to define the UNET Model"""
    # Contracting Path
    c1 = conv2d_block(input_img, n_filters * 1, kernel_size=3, batchnorm=batchnorm)
    p1 = tf.keras.layers.MaxPooling2D((2, 2))(c1)
    # p1 = tf.keras.layers.Dropout(dropout)(p1)

    c2 = conv2d_block(p1, n_filters * 2, kernel_size=3, batchnorm=batchnorm)
    p2 = tf.keras.layers.MaxPooling2D((2, 2))(c2)
    # p2 = tf.keras.layers.Dropout(dropout)(p2)

    c3 = conv2d_block(p2, n_filters * 4, kernel_size=3, batchnorm=batchnorm)
    p3 = tf.keras.layers.MaxPooling2D((2, 2))(c3)
    # p3 = tf.keras.layers.Dropout(dropout)(p3)

    c4 = conv2d_block(p3, n_filters * 8, kernel_size=3, batchnorm=batchnorm)
    p4 = tf.keras.layers.MaxPooling2D((2, 2))(c4)
    # p4 = tf.keras.layers.Dropout(dropout)(p4)

    c5 = conv2d_block(p4, n_filters * 16, kernel_size=3, batchnorm=batchnorm)
    p5 = tf.keras.layers.MaxPooling2D((2, 2))(c5)
    # p5 = tf.keras.layers.Dropout(dropout)(p5)

    c6 = conv2d_block(p5, n_filters=n_filters * 32, kernel_size=3, batchnorm=batchnorm)

    # Expansive Path
    u7 = tf.keras.layers.Conv2DTranspose(n_filters * 16, (3, 3), strides=(2, 2), padding='same')(c6)
    u7 = tf.keras.layers.concatenate([u7, c5])
    u7 = tf.keras.layers.Dropout(dropout)(u7)
    c7 = conv2d_block(u7, n_filters * 16, kernel_size=3, batchnorm=batchnorm)

    u8 = tf.keras.layers.Conv2DTranspose(n_filters * 8, (3, 3), strides=(2, 2), padding='same')(c7)
    u8 = tf.keras.layers.concatenate([u8, c4])
    u8 = tf.keras.layers.Dropout(dropout)(u8)
    c8 = conv2d_block(u8, n_filters * 8, kernel_size=3, batchnorm=batchnorm)

    u9 = tf.keras.layers.Conv2DTranspose(n_filters * 4, (3, 3), strides=(2, 2), padding='same')(c8)
    u9 = tf.keras.layers.concatenate([u9, c3])
    u9 = tf.keras.layers.Dropout(dropout)(u9)
    c9 = conv2d_block(u9, n_filters * 4, kernel_size=3, batchnorm=batchnorm)

    u10 = tf.keras.layers.Conv2DTranspose(n_filters * 2, (3, 3), strides=(2, 2), padding='same')(c9)
    u10 = tf.keras.layers.concatenate([u10, c2])
    u10 = tf.keras.layers.Dropout(dropout)(u10)
    c10 = conv2d_block(u10, n_filters * 2, kernel_size=3, batchnorm=batchnorm)

    u11 = tf.keras.layers.Conv2DTranspose(n_filters * 1, (3, 3), strides=(2, 2), padding='same')(c10)
    u11 = tf.keras.layers.concatenate([u11, c1])
    u11 = tf.keras.layers.Dropout(dropout)(u11)
    c11 = conv2d_block(u11, n_filters * 1, kernel_size=3, batchnorm=batchnorm)

    outputs = tf.keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c11)
    model = tf.keras.models.Model(inputs=[input_img], outputs=[outputs])
    return model
```

```python
# Build U-Net model with specified parameters
model_unet = get_unet(
    input_img, 
    n_filters=16, 
    dropout=0.05, 
    batchnorm=True
)

# Compile the model with appropriate loss function for binary segmentation
model_unet.compile(
    optimizer='adam', 
    loss='binary_crossentropy', 
    metrics=['accuracy']
)
```
